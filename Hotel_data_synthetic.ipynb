```python
from nltk.metrics.distance import jaccard_distance
```

```python
list1 = scrape_hotel_list
list2 = hotel_list
```

```python
type(list2[0])
```

```python
list1 = list(map(str, list1))
list2 = list(map(str, list2))
```

```python
def jaccard_similarity(sentence1, sentence2):
    return 1 - jaccard_distance(set(sentence1.split()), set(sentence2.split()))
```

```python
similar_sentences = []
threshold = 0.5
for sentence1 in list1:
    for sentence2 in list2:
        similarity = jaccard_similarity(sentence1, sentence2)
        if similarity > threshold:
            similar_sentences.append((sentence1, sentence2, similarity))
```

```python
df = pd.DataFrame(similar_sentences,columns=['Scrape_hotel_Name','Hotel_name','Similarity'])
```
